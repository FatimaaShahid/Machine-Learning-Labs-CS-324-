```python
#float to int; decimal point values will be truncated
arr = np.array([3.7, -1.2, -2.6, 0.5, 12.9, 10.1])
print(arr)
new = arr.astype(np.int32)
new
```

```python
#we can match the data type as per another array as well
int_array = np.arange(10)
calibers = np.array([.22, .270, .357, .380, .44, .50],
dtype=np.float64)
match = int_array.astype(calibers.dtype)
print(int_array.dtype)
print(calibers.dtype)
match.dtype
```
